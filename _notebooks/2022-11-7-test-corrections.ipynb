## Question 1

```
public static int myster(int[] arr) {
    int x = 0;
    for (int k=0; k<arr.length; k=k+2) {
        x = x + arr[k];
    }
    return x;
}
```

Assume that the array nums has been declared and initialized as follows.

int [ ] nums = { 3, 6, 1, 0, 1, 4, 2};

What value will be returned as a result of the call mystery(nums) ?


### My Answer
B) 6

### Actual Answer
C) 7

The for loop initializes ``k`` at 0. It will increment by 2 until it is greater than or equal to ``arr.length``. k will take on the values of 0, 2, 3, and 6 until the loop stops. The loops will continue iterating until k is 6, at which ``x+arr[k]`` is 7.
